# Data Preprocessing & Cleaning

```python
# kolom drop: Ingredients, Steps, URL, Title, 
menu_fix = menu[['Title','Ingredients','Steps','URL', 'Title Cleaned', 'Category', 'Ingredients Cleaned', 'Total Ingredients', 'Total Steps', 'Loves']]
menu_fix.head() 
```

```python
menu_fix.isnull().sum()
```

```python
menu_fix.dropna(inplace=True)
```

```python
menu_fix.isnull().sum()
```

```python
menu_fix.duplicated(subset='Title Cleaned').sum()
```

```python
menu_fix = menu_fix.drop_duplicates(subset='Title Cleaned', keep='first')
```

```python
menu_fix.shape
```

```python
menu_fix['Loves'].unique()
```

```python
# Buat function untuk mengubah loves menjadi rating
def loves_to_rating(loves):
    if loves < 5:
        return 1
    elif loves < 10:
        return 2
    elif loves < 30:
        return 3
    elif loves < 100:
        return 4
    else:
        return 5

menu_fix['Rating'] = menu_fix['Loves'].apply(loves_to_rating)
```

```python
menu_fix = menu_fix.drop(columns=['Loves'])
menu_fix
```

```python
menu_fix['Category'] = menu_fix['Category'].str.capitalize()
```

```python
# Buat kolom function untuk menggabungkan kolom total ingredients dan total steps
# Setelah itu buat kolom baru bernama 'Complexity'

def classify_complexity(steps, ingredients):
    if ingredients <= 9 and steps <= 4:
        return 'Cepat & Mudah'
    elif ingredients <= 15 and steps <= 7:
        return 'Butuh Usaha'
    else:
        return 'Level Dewa Masak'

menu_fix['Complexity'] = menu_fix.apply(lambda x: classify_complexity(x['Total Steps'], x['Total Ingredients']), axis=1)
menu_fix = menu_fix.drop(columns=['Total Ingredients', 'Total Steps'])
menu_fix
```

```python
menu_fix['tags'] = menu_fix['Title Cleaned'] + ', ' + menu_fix['Category'] + ', ' + menu_fix['Ingredients Cleaned'] + menu_fix['Complexity']
menu_fix_banget = menu_fix.drop(columns=['Title Cleaned', 'Ingredients Cleaned', 'Category', 'Complexity'])
menu_fix_banget
```

# Stemming dengan Sastrawi

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

menu_fix_banget['tags_stemmer'] = menu_fix_banget['tags'].apply(lambda x: ' '.join([stemmer.stem(word) for word in x.split(', ')]))
```

# TF-IDF Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(menu_fix_banget['tags_stemmer'])
```

# Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix)
```

# Fungsi Rekomendasi

```python
def recommend_from_keyword(ingredients_list, top_n=10, min_rating=None, max_rating=None, complexity_filter=None):
    # Gabungkan semua bahan jadi satu string
    combined_ingredients = ' '.join(ingredients_list)
    stemmed_ingredients = ' '.join([stemmer.stem(word) for word in combined_ingredients.split()])
    
    # 2. TF-IDF vector untuk keyword
    keyword_vector = vectorizer.transform([stemmed_ingredients])
    
    # 3. Hitung similarity
    sim_scores = cosine_similarity(keyword_vector, tfidf_matrix).flatten()
    
    # 4. Ambil top_n * 2 hasil teratas (biar bisa difilter)
    top_indices = sim_scores.argsort()[::-1]
    
    results = menu_fix.iloc[top_indices].copy()
    results['similarity'] = sim_scores[top_indices]
    
    # 5. Filter opsional
    if min_rating is not None:
        results = results[results['Rating'] >= min_rating]
    if max_rating is not None:
        results = results[results['Rating'] <= max_rating]
    if complexity_filter:
        results = results[results['Complexity'] == complexity_filter]
    
    # 6. Return hasil top_n
    return results[['Title','Category','Ingredients', 'Steps', 'URL', 'Rating', 'Complexity']].head(top_n)
```

```python
# Contoh
Top_10_olahan_telor_simple = recommend_from_keyword(
    ingredients_list=['Telur', 'Ayam'],
    top_n=10,
    complexity_filter='Cepat & Mudah',
    min_rating=4,
    max_rating=5
)
Top_10_olahan_telor_simple
```

```python
# Jika ada kolom penghubung, misalnya 'Title'
merged_df = menu_fix.merge(menu_fix_banget[['Title', 'tags_stemmer']], on='Title', how='left')
```

```python
merged_df.head()
```

```python
import pickle
import json
import numpy as np
import pandas as pd
from scipy.sparse import save_npz, load_npz
import gzip
import os

# %%
print("Menyimpan model dan data dengan optimisasi...")

# Buat direktori models jika belum ada
os.makedirs('models', exist_ok=True)

# %%
# 1. Simpan TF-IDF Vectorizer (tetap dalam format pickle karena ukurannya kecil)
print("Saving TF-IDF Vectorizer...")
with open('models/tfidf_vectorizer.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)

# %%
# 2. OPTIMISASI: Simpan hanya data penting dari Cosine Similarity Matrix
# Alih-alih menyimpan matrix penuh, kita simpan dalam format sparse dan compressed
print("Saving optimized Cosine Similarity Matrix...")

# Konversi ke sparse matrix (hanya simpan nilai > threshold tertentu)
threshold = 0.1  # Hanya simpan similarity > 0.1
cosine_sim_sparse = cosine_sim.copy()
cosine_sim_sparse[cosine_sim_sparse < threshold] = 0

# Simpan dalam format compressed sparse
from scipy.sparse import csr_matrix
cosine_sparse = csr_matrix(cosine_sim_sparse)
save_npz('models/cosine_similarity_sparse.npz', cosine_sparse)

print(f"Original cosine matrix size: {cosine_sim.nbytes / (1024*1024):.2f} MB")
print(f"Sparse matrix file size: {os.path.getsize('models/cosine_similarity_sparse.npz') / (1024*1024):.2f} MB")

# %%
# 3. Simpan DataFrame dalam format compressed CSV
print("Saving processed recipes...")
menu_fix.to_csv('models/processed_recipes.csv.gz', index=False, compression='gzip')

# %%
# 4. Simpan stemmer (ukuran kecil, keep as is)
print("Saving stemmer...")
with open('models/stemmer.pkl', 'wb') as f:
    pickle.dump(stemmer, f)

# %%
# 5. OPTIMISASI: Simpan data minimal dalam JSON compressed
def save_optimized_json():
    print("Creating optimized JSON files...")
    
    # Hanya simpan data essential untuk recipes
    recipes_json = []
    for idx, row in merged_df.iterrows():
        recipe = {
            'id': str(idx),
            'title': row['Title'],
            'category': row['Category'],
            'rating': int(row['Rating']),
            'complexity': row['Complexity'],
            'tags': row['tags']
        }
        # Simpan ingredients dan steps dalam format compact
        if isinstance(row['Ingredients'], str):
            recipe['ingredients'] = row['Ingredients']
        if isinstance(row['Steps'], str):
            recipe['steps'] = row['Steps']
        if isinstance(row['URL'], str):
            recipe['url'] = row['URL']
            
        recipes_json.append(recipe)
    
    # Simpan dalam format compressed JSON
    with gzip.open('models/recipes.json.gz', 'wt', encoding='utf-8') as f:
        json.dump(recipes_json, f, ensure_ascii=False, separators=(',', ':'))
    
    # Simpan vocabulary dalam format minimal
    tfidf_vocab = {
        'vocabulary': vectorizer.vocabulary_,
        'feature_names': vectorizer.get_feature_names_out().tolist()
    }
    
    with gzip.open('models/tfidf_vocab.json.gz', 'wt', encoding='utf-8') as f:
        json.dump(tfidf_vocab, f, separators=(',', ':'))
    
    print("Optimized JSON files created!")

# %%
# 6. ELIMINASI: Skip cosine_matrix.json karena terlalu besar
# Kita akan hitung similarity on-the-fly menggunakan TF-IDF
print("Skipping large cosine_matrix.json - will compute on-the-fly")

# Jalankan optimisasi
save_optimized_json()

# %%
# 7. Print ukuran file untuk monitoring
print("\n=== FILE SIZES ===")
model_files = [
    'models/tfidf_vectorizer.pkl',
    'models/cosine_similarity_sparse.npz',
    'models/processed_recipes.csv.gz',
    'models/stemmer.pkl',
    'models/recipes.json.gz',
    'models/tfidf_vocab.json.gz'
]

total_size = 0
for file_path in model_files:
    if os.path.exists(file_path):
        size_mb = os.path.getsize(file_path) / (1024*1024)
        print(f"{file_path}: {size_mb:.2f} MB")
        total_size += size_mb
    else:
        print(f"{file_path}: FILE NOT FOUND")

print(f"\nTotal model size: {total_size:.2f} MB")

if total_size > 900:  # Buffer 100MB dari limit 1GB
    print("⚠️  WARNING: Total size might be too large for HF Spaces!")
    print("Consider further optimization or use Git LFS")
else:
    print("✅ Model size is suitable for HF Spaces")

print("\nModel optimization completed!")

# %%
# 8. Optional: Hapus file besar yang tidak diperlukan
print("\nCleaning up large unnecessary files...")
files_to_remove = [
    'models/cosine_matrix.json',  # File ini sangat besar
    'models/cosine_similarity_matrix.pkl'  # Kita sudah punya versi sparse
]

for file_path in files_to_remove:
    if os.path.exists(file_path):
        size_mb = os.path.getsize(file_path) / (1024*1024)
        os.remove(file_path)
        print(f"Removed {file_path} ({size_mb:.2f} MB)")

print("Cleanup completed!")
```
